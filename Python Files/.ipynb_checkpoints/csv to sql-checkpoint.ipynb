```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Data Files\\Pollution Data.csv')

print(df.shape)
df.head()
```

```python
li_df_columns = list(df.columns)

li_df_columns
```

```python
df.rename(columns={"ISO 3166-1 alpha-3" : "Country Code"}, inplace=True)
```

```python
dict_df_columns = {"Country" : "VARCHAR(50)", "Country Code" : "VARCHAR(5)", "Year" : "INT", "Total" : "Float", 
                   "Coal" : "Float", "Oil" : "Float", "Gas" : "Float", "Cement" : "Float", "Flaring" : "Float", 
                   "Other" : "Float", "Per Capita" : "Float"}
dict_df_columns
```

```python
s = f"CREATE TABLE pollution(\n"

for i, k in zip(range(len(dict_df_columns)), dict_df_columns.keys()):
    s += f"{k}\t\t{dict_df_columns[k]}"
    
    if i != len(dict_df_columns) - 1:
        s += ",\n"
    else:
        s += ");"


print(s)
```

```python
df.describe(include="all")
```

```python
s = f"INSERT INTO pollution VALUES "

start = 30000
end = 30004

for i in range(start, end):
    
    ser = df.iloc[i]
    s += "("
    
    for j in range(len(dict_df_columns)):
        
        if pd.isna(ser[j]):
            s += "NULL"
        elif j > 1:
            s += f"{ser[j]}"
        else :
            s += f"'{ser[j]}'"
        
        if j != len(dict_df_columns) - 1:
            s += ", "
    
    if i != end - 1:
        s += "),\n"
    else:
        s += ");"
        
print(s)
```

```python
s = f"INSERT INTO pollution VALUES "

start = 0
end = 272

for i in range(start, end):
    
    ser = df[df.Country == "Côte d'Ivoire"].iloc[i]
    s += "("
    
    for j in range(len(dict_df_columns)):
        
        if pd.isna(ser[j]):
            s += "NULL"
        elif j > 1:
            s += f"{ser[j]}"
        elif j == 0:
            s += f"'Côte d Ivoire'"
        else:
            s += f"'CIV'"
        
        if j != len(dict_df_columns) - 1:
            s += ", "
        
    if i != end - 1:
        s += "),\n"
    else:
        s += ");"
        
print(s)
```
